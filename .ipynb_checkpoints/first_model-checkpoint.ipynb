```python
import pandas as pd
import numpy as np 
import IPython
from scipy.io import wavfile
import matplotlib.pyplot as plt
import librosa
import librosa.display
from playsound import playsound
import os
import sklearn
```

```python
# take the wav files; generate the numeric features 
wav_file_1 = "sample_set/overlap/audio_chunk_3.0.wav"
x , sr = librosa.load(wav_file_1)

mfccs = librosa.feature.mfcc(x, sr=sr)
print(mfccs.shape)
#Displaying  the MFCCs:
librosa.display.specshow(mfccs, sr=sr, x_axis='time')

# basic classifier model with sklearn 
```

```python
mfcc_features = []
for filename in os.listdir("sample_set/overlap"):
    if filename.endswith(".wav"): 
        print(filename)
        x , sr = librosa.load("sample_set/overlap/" + filename)
        mfccs = librosa.feature.mfcc(x, sr=sr)
        mfccs = sklearn.preprocessing.scale(mfccs, axis=1)
        print(mfccs)
        mfcc_features.append((mfccs,1))
        
for filename in os.listdir("sample_set/no_overlap"):
    if filename.endswith(".wav"): 
        print(filename)
        x , sr = librosa.load("sample_set/no_overlap/" + filename)
        mfccs = librosa.feature.mfcc(x, sr=sr)
        mfccs = sklearn.preprocessing.scale(mfccs, axis=1)
        print(mfccs)
        mfcc_features.append((mfccs,0))
    
    

```

```python
mfcc_features = list(zip(*mfcc_features))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

# mfccs = sklearn.preprocessing.scale(mfccs, axis=1)
model = KNeighborsClassifier(n_neighbors=3)

features = [x.flatten() for x in mfcc_features[0]]
labels = list(mfcc_features[1])

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=1, stratify=labels)

model.fit(X_train,y_train)
```

```python
print(model.predict(X_test))
y_test
```
