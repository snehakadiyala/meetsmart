```python
mfcc_features = []
for filename in os.listdir("sample_set/overlap"):
    if filename.endswith(".wav"): 
        print(filename)
        x , sr = librosa.load("sample_set/overlap/" + filename)
        mfccs = librosa.feature.mfcc(x, sr=sr)
        mfccs = sklearn.preprocessing.scale(mfccs, axis=1)
        print(mfccs)
        mfcc_features.append((mfccs,1))
        
for filename in os.listdir("sample_set/no_overlap"):
    if filename.endswith(".wav"): 
        print(filename)
        x , sr = librosa.load("sample_set/no_overlap/" + filename)
        mfccs = librosa.feature.mfcc(x, sr=sr)
        mfccs = sklearn.preprocessing.scale(mfccs, axis=1)
        print(mfccs)
        mfcc_features.append((mfccs,0))
    
    

```
